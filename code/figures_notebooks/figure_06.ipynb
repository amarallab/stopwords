## get all data

```python
N_s = 1000
corpus_name = 'PBMC_1k_RNAseq_rawcount'

path_read = os.path.join(os.pardir,'output')
fname_read = '%s_stopword-statistics_Ns%s.csv'%(corpus_name,N_s)
filename = os.path.join(path_read,fname_read)
df = pd.read_csv(filename,index_col=0)
```

## sort the data

```python
df = df.sort_values(by='N', ascending=0)
```

```python
word_count_array = df['N'].values

info_mean_array = df['H-tilde'].values
info_real_array = df['H'].values
yerr_array = df['H-tilde_std'].values * 2
```

```python
chosen_word_list = ['PTGDS','GNLY', 'HLA-A', 'MALAT1']

chosen_word_dict = {}

for i in chosen_word_list:

    chosen_word_dict[i] = {
        'word_count': float(df.loc[i]['N']),
        'info_real': float(df.loc[i]['H']),
    }

chosen_word_dict
```

```python
x_factor_list = [.9, .9, .2,  .5,]
y_factor_list = [-1.0, -1.0, 0.7, -1.1,]
chosen_word_list
```
